```python
X = df.iloc[:,[0,1,2,3]]
y = df['species']
X.head()
```

```python
y[:5]
```

#KNN (K-Nearest Neighbour) METHOD

```python
### Apply knn for IRIS Dataset
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors =3)
knn
```

```python
knn.fit(X,y)
```

```python
y_prediction = knn.predict(X)
y_prediction
```

#Predicting Accuracy in KNN method 1

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_prediction)
```

```python
knn1 = KNeighborsClassifier(n_neighbors = 5) # if n_neighbors = 10....accuracy_score=0.98
knn1
```

```python
knn1.fit(X,y)
```

```python
y_prediction1 = knn1.predict(X)
```

##Predicting Accuracy in KNN method 2


```python
accuracy_score(y,y_prediction1)
```

```python
knn1.predict([[3.1,3.5,5.4,0.2]])
```

##SVM (Support Vector Machine) MODEL

```python
svm = SVC()
svm
```

```python
svm.fit(X,y)
```

```python
y_prediction2 = svm.predict(X)
y_prediction2
```

##predicting accuracy in SVM method 1
